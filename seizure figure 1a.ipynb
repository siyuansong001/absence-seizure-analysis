```python
import scanpy as sc
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib import colors
import glob
import anndata
import sys
from scipy.io import mmread
sys.path.append(r'/media/data2/samba/xjlab/')
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor='white')
```

```python
adata = sc.read_h5ad('/media/data2/samba/ssong/data/S1/merged V1 and ALM.h5ad')
```

```python
import matplotlib.pyplot as plt

custom_palette = {
    'ExNs': '#1f77b4',  
    'Lamp5': '#ff7f0e',  
    'Non-neuron': 'gray',  
    'Pvalb Akr1c18 Ntf3': '#2ca02c',  
    'Pvalb Calb1 Sst': '#d62728',  
    'Pvalb Gabrg1': '#9467bd',  
    'Pvalb Gpr149 Islr': '#8c564b',  
    'Pvalb Reln Itm2a': '#e377c2',  
    'Pvalb Reln Tac1': '#9500ff',  
    'Pvalb Sema3e Kank4': '#bcbd22',  
    'Pvalb Th Sst': '#17becf',  
    'Pvalb Tpbg': '#8c6bb1',  
    'Pvalb Vipr2': '#ffbb78',  
    'Sst': '#98df8a',  
    'Vip': '#ffbb00'  
}

adata.uns.pop('celltype2_colors', None) 
adata.uns['celltype2_colors'] = [
    custom_palette[cluster] for cluster in adata.obs['celltype2'].cat.categories
]

sc.pl.tsne(
    adata,
    color='celltype2',
    palette=[custom_palette[c] for c in adata.obs['celltype2'].cat.categories],  
    legend_fontsize=11,
    legend_fontweight='light',
    legend_fontoutline=0,
    title="t-SNE with Corrected Vip Color", save='cluster333.svg'
)
```
